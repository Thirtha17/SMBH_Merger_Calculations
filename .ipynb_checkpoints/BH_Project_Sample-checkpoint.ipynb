```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import numpy
import h5py

#Seaborn is used for one alternate plotting method at the end
#If you don't have it installed, you can comment this out, along with the cells which do Seaborn plots (anywhere there is something that has sns.)
import seaborn as sns


```

```python
#Define parameters from Illustris simulation
h=0.6774
boxlen=75.0/h #in Mpc
```

```python
#Open blackhole_mergers.hdf5 file
illustris_mergerdata=h5py.File('blackhole_mergers.hdf5', 'r')
#Print list of data stored in blackhole_mergers file
list(illustris_mergerdata.keys())
```

```python
print(illustris_mergerdata['mass_in'].value)
print(illustris_mergerdata['mass_out'].value)
```

```python
plt.scatter(np.log10(illustris_mergerdata['mass_in']), np.log10(illustris_mergerdata['mass_out']))
```

```python
#Define 1D arrays for the data used in this notebook (scale factor, redshift, and masses)
mergea=illustris_mergerdata['time'][:,]
mergez=1.0/mergea-1.0
mergemin=illustris_mergerdata['mass_in'][:,]*1e10/h
mergemout=illustris_mergerdata['mass_out'][:,]*1e10/h
```

```python
#Create arrays of M1 and M2
#    M1: Mass of more massive black hole in the merger
#    M2: Mass of less massive black hole in the merger
mergem1=numpy.maximum(mergemin, mergemout)
mergem2=numpy.minimum(mergemin, mergemout)
```

```python
plt.scatter(np.log10(mergem1), np.log10(mergem2))
```

```python
#Plot 1D histogram of redshift distribution of mergers
plt.hist(mergez, histtype='step')
plt.xlabel('z')
plt.ylabel('Number')
```

```python
#Plot 2-panel histogram of the logarithmic merger mass ratio
#   Both panels show the same information, just presented differently
#   Top panel uses linear y axis
#   Bottom panel uses logarithmic y axis
fig,axs=plt.subplots(2, sharex=True, sharey=False, gridspec_kw={'hspace':0})
axs[0].hist(numpy.log10(mergem2/mergem1), histtype='step', color='blue')
axs[1].hist(numpy.log10(mergem2/mergem1), histtype='step', color='blue')
axs[1].set_yscale('log')
axs[1].set_xlabel('Log(M$_2$/M$_1$)')
```

```python
#Plot histogram comparing distributions of M1 and M2
#     Here we plot the number of mergers per cubic megaparsec (comoving)
#     Note we also normalize by the width of the bins, so different binsizes can be used without biasing results
def histcompare(x, histlabel, binwidth):
    binmin=min(x)
    binmax=max(x)
    binlist=[binmin+x*binwidth for x in range(int(numpy.ceil((binmax-binmin)/binwidth)))]
    plt.hist(x, histtype='step', label=histlabel, bins=binlist, weights=numpy.ones_like(x)/numpy.power(boxlen,3)/binwidth)
    
plt.close()
histcompare(numpy.log10(mergem1), 'M$_1$', 0.1)
histcompare(numpy.log10(mergem2), 'M$_2$', 0.1)
plt.ylabel('dN/dlogM')
plt.xlabel('log(M/M$_\odot$)')
plt.legend()
```

```python
#2D histogram comparing M1 and M2
plt.hist2d(numpy.log10(mergem1), numpy.log10(mergem2), bins=50)
plt.xlabel("M$_1$")
plt.ylabel("M$_2$")
plt.colorbar()
plt.show()
```

```python
#Sample calculations using AstroPy
import astropy
from astropy.cosmology import FlatLambdaCDM
import astropy.units as u
from astropy.cosmology import z_at_value

#Can call a built-in cosmology
#from astropy.cosmology import WMAP9 as wmap9

#Can defint our own cosmology parameters
cosmo=FlatLambdaCDM(H0=100*h, Om0=.2726, Ob0=0.0456)
```

```python
cosmo.age(4)
```

```python
#Calculate age of universe at several redshifts:
zlist=[0,1,2]
print('Age of universe at redshifts', zlist, ':')
cosmo.age(zlist)
```

```python
#Calculate redshift at specified age of the universe:
z_at_value(cosmo.age, 7*u.Gyr)
```

```python
#Calculate luminosity distance at specified redshift:
cosmo.luminosity_distance(3)
```

```python
#Note that most of these commands return a quantity which includes units
#To get numerical value or units separately, use the following:
cosmo.age(zlist).value, cosmo.age(zlist).unit
```

```python
#3 Methods for comparing data between 2D histograms:

#Method 1: Overplot contours using Seaborn's KDEPlot
fig, ax=plt.subplots()
sns.kdeplot(numpy.log10(mergem1), numpy.log10(mergem2), ax=ax, color='blue')
sns.kdeplot(numpy.log10(mergem1*3), numpy.log10(mergem2+3e5), ax=ax, color='green')
```

```python
#Method 2: Calculate 2D histograms, and plot their contours using PyPlot
#     KDEPlot is slower, but does smoothing calculations on the underlying data; hence this looks noisier
H, xedges, yedges=numpy.histogram2d(numpy.log10(mergem1), numpy.log10(mergem2), bins=20)
H=H.T
xbins=xedges[:-1] + (xedges[1] - xedges[0]) / 2
ybins = yedges[:-1] + (yedges[1] - yedges[0]) / 2
plt.contour(xbins,ybins,H, colors='blue')

H2, xedges2, yedges2=numpy.histogram2d(numpy.log10(mergem1*3), numpy.log10(mergem2+3e5), bins=20)
H2=H2.T
xbins2=xedges2[:-1] + (xedges2[1] - xedges2[0]) / 2.
ybins2 = yedges2[:-1] + (yedges2[1] - yedges2[0]) / 2.
plt.contour(xbins2,ybins2,H2, colors='green')
```

```python
#TEST: Overplot KDEPlot and by-hand contouring, showing consistent results
fig, ax=plt.subplots()
ax.contour(xbins,ybins,H, colors='green')
sns.kdeplot(numpy.log10(mergem1), numpy.log10(mergem2), ax=ax, color='blue')
```

```python
#Method 3: Calculate 2D histogram data for each sample, and plot ratio or difference between histograms
#      Note: This lets you explicitly compare datasets.  However, it is easy to miss data if you don't 
#                      carefully consider everything
#           e.g. For difference plot you must consider both negative and positive values.
#                For ratio plot, you must consider both 0 and NaN values (i.e. when one dataset has 0 in a given bin)
#RECOMMENDATION: Use contour plotting (Method 1 or 2) to get general comparison.  Then if you want to use this in
#                      addition, you can try it out

#First define edges of histogram bins
xedges=[5+0.1*x for x in range(55)]
yedges=[5+0.1*x for x in range(55)]
#Compute 2D histogram data for first dataset
H, xedges, yedges=numpy.histogram2d(numpy.log10(mergem1), numpy.log10(mergem2), bins=[xedges,yedges])
H=H.T
X,Y=numpy.meshgrid(xedges, yedges)

#Compute 2D histogram data for second dataset
H2, xedges, yedges=numpy.histogram2d(numpy.log10(mergem1*3), numpy.log10(mergem2), bins=[xedges,yedges])
H2=H2.T
X2,Y2=numpy.meshgrid(xedges, yedges)

#Plot 2D histogram of ratio between two datasets
im=plt.pcolormesh(X,Y,H2/H)
#Add a colorbar showing the values corresponding to the given colors
plt.colorbar(im)

#Analysis: 
#   Rightmost bins show the largest increase, where highest-M1 mergers occur
#   Also see that we completely lose all mergers along the main diagonal (by increasing M1)
```

```python
#As above, but plotting the difference in merger number, instead of ratio

from matplotlib.colors import LogNorm

fig, ax=plt.subplots()
#Here I plot using the 'seismic' color map
#     I also specify the range for the colorbar, so we see increases in red and decreases in blue
im=ax.pcolormesh(X,Y,H2-H,cmap=plt.get_cmap('seismic'), norm=matplotlib.colors.Normalize(vmin=-1*numpy.max(numpy.abs(H2-H)),vmax=numpy.max(numpy.abs(H2-H))))
fig.colorbar(im, ax=ax)

#Analysis:
#    Again we see that rightmost edge sees increase.  However, it's a relatively small absolute number
#             (But the small number of mergers is what caused the larger values in the ratio plot above)
#    Here we see that there is a significant increase in mergers at [M1~2e7, M2~2e6], where we have shifted the peak from the original distribution
#             (But the high number of mergers in those bins results in the small values in ratio plot above)
#    We also see the decrease along the main diagonal, as seen in the ratio plot above
```

```python
#Outline for creating/using Lookup Table:
#    Note that I use very few points here so you can see how it works, but you should use a larger number for your analysis
```

```python

#Make array of times which we will use for the lookup table
t_list=numpy.linspace(1,13,10)

#Make array of redshifts corresponding to those times
z_list=[]
for i in t_list:
    z_list.append(z_at_value(cosmo.age, i*u.Gyr))
    
#Plot the points to see what our lookup table looks like:
plt.scatter(z_list, t_list)

print(t_list)
```

```python
#Here we will interpolate by hand
#    Note: You can take advantage of interpolation commands within python (see next cell) to avoid doing all of this
#          However, you should make sure you understand what is being done, which this might help you with

#Select a target age that you want to know the redshift at:
target_t=1.5

#Interpolate between the two closest points, to estimate the redshift:

#Find array location with larger age than target
#   Note that bisect requires the array be sorted
import bisect
interp1=bisect.bisect(t_list, target_t)
#And thus also the array location with smaller age than target (given th array is sorted)
interp2=interp1-1

#Find redshift using interpolation between those two points
found_z=z_list[interp1]+(z_list[interp2]-z_list[interp1])*(target_t-t_list[interp1])/(t_list[interp2]-t_list[interp1])

#Plot interpolated point for comparison
plt.scatter(z_list, t_list, color='blue')
plt.scatter([found_z], [target_t], color='red')
#plt.plot(z_list[interp2:interp1+1], t_list[interp2:interp1+1])
```

```python
#Instead of doing interpolation by hand, we can take advantage of the numpy.interp routine
#   Much less to write, but gives the same result
#   Note there are also other interpolation approaches you can use as well (e.g. scipy.interpolate)
found_z_numpy=numpy.interp(target_t, t_list, z_list)

print("Interpolation by hand:", found_z)
print("Using numpy.interp:   ", found_z_numpy)

#plt.scatter(z_list, t_list, color='blue')
#plt.scatter(found_z_numpy, [target_t], color='green')
```

```python
#We can also pass a full array into numpy.interp, instead of a single value:
#Here we define list of times (target_tlist) and calculate the redshifts associated with them (interp_zlist)
target_tlist=numpy.linspace(1.5,12, 100)
interp_zlist=numpy.interp(target_tlist, t_list, z_list)

plt.scatter(z_list, t_list, color='blue')
plt.plot(interp_zlist, target_tlist, color='red')

#In blue are the points from our lookup table
#In red are the interpolated redshifts (interp_zlist) calculated from list of ages (target_tlist)
```
